## Smooth Support Vector Classifier

#### 1. generate data

```python
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

```python
point_num = 100
feature_num = 2
data_size = np.ones(shape=(point_num, feature_num))
```

```python
x_pos = np.random.normal(1.5*data_size)
y_pos = np.ones(point_num)
```

```python
x_neg = np.random.normal(1.5*-data_size)
y_neg = np.full((point_num), -1)
```

```python
X = np.r_[x_pos, x_neg]
y = np.r_[y_pos, y_neg]
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

#### 2.SSVC
- X: shape must be (m, n), which m is data size and n is feature number.
- y: shape must be (1, m) or (m, 1), which m is data size.

```python
from ssvm import SSVC
svc = SSVC()
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
```

```python
svc.fit(X_train, y_train)
```

```python
svc.print_params()
```

```python
svc.score(X_val, y_val)
```

```python
prediction = svc.predict(X_val)
```

```python
plt.scatter(X_val[:, 0], X_val[:, 1], c=prediction)
```
